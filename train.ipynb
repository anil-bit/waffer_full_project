#apply cluster to your data

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42) # initializing the KMeans object
    kmeans.fit(x) # fitting the data to the KMeans Algorithm
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11), wcss)
plt.xlabel("number of cluster")
plt.ylabel("wcss")
plt.show()
```

#there are two ways to get the cluster 
#first one is to get the no of cluster using graph
#second one is get the no of cluster using knee locater function

```python
from kneed import KneeLocator
kn = KneeLocator(range(1, 11), wcss, curve='convex', direction='decreasing')
```

```python
kn.knee
```

```python
number_of_clusters = kn.knee
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics  import roc_auc_score,accuracy_score
from sklearn.ensemble import RandomForestClassifier



class model_finder:
    #this class will find the best model 
    def __init__(self):
        self.clf = RandomForestClassifier()
    
    def get_best_params_for_xgboost(self,train_x,train_y):
        #this method gets best parameter for xg_boost
        self.param_grid_xg_boost = {"learning_rate":[0.5,0.1,0.01,0.001],
                                    "max_depth":[3,5,10,20],
                                    "n_estimator":[10,20,100,200]
                                   }
        #craeting an object sor grid search 
        self.grid_search = GridSearchCV(XGBClassifier(objective='binary:logistic'),self.param_grid_xg_boost, verbose=3,cv=5) 
        self.grid_search.fit(train_x,train_y)
        
        #extracting the best parameter
        self.learning_rate = self.grid_search.best_params_["learning_rate"]
        self.max_depth = self.grid_search.best_params_["max_depth"]
        self.n_estimator =self.grid_search.best_params_["n_estimator"]
        
        self.xgb = XGBClassifier(learning_rate=self.learning_rate,max_depth = self.max_depth,n_estimators=self.n_estimator)
        self.xgb.fit(train_x,train_y)
        return self.xgb
    
    
    def get_best_prams_for_random_forest(train_x,train_y):
        #get best random forest parameters 
        self.param_grid_random_forest ={"n_estimator":[10, 50, 100, 130],
                                        "criterion":['gini', 'entropy'],
                                        "max_depth":range(2, 4, 1),
                                        "max_features": ['auto', 'log2']}
        

        
        self.grid_search = GridSearchCV(estimator=self.clf,param_grid=self.param_grid_random_forest, cv=5, verbose=3)
        self.grid_search.fit(train_x,train_y)
        
        #extracting the best parameters
        self.n_estimator = self.grid_search.best_params_["n_estimator"]
        self.criterion = self.grid_search.best_params_["criterion"]
        self.max_depth = self.grid_search.best_params_["max_depth"]
        self.max_features = self.grid_search.best_params_["max_features"]
        
        #creating a new model with the best parameter
        self.clf = RandomForestClassifier(n_estimators=self.n_estimators, criterion=self.criterion,
                                              max_depth=self.max_depth, max_features=self.max_features)
        self.clf.fit(train_x,train_y)
        return self.clf
        
    
    
    
    
    def get_best_model(self,train_x,train_y,test_x,test_y):
        
        #get the best model
        #create the best model for xg_boost 
        self.xgboost = self.get_best_params_for_xgboost(train_x,train_y)
        self.predict_xgboost = self.xgboost.predict(test_x)#prediction using the xgboost model
        
        if len(test_y.unique()) == 1:#if there is only one label in y,then we cdompare the accuraccy of the model
            self.xgboost_score = accuracy_score(test_y,self.predict_xgboost)
        else:
            self.xgboost_score = roc_auc_score(test_y,self.predict_xgboost)#auc for xgboost
            
            
        #create the best model for random_forest
        
        self.random_forest = self.get_best_prams_for_random_forest(train_x,train_y)
        self.predict_random = self.random_forest.predict(test_x)
        
        if len(test_y.unique()) == 1:#if there is only one label in y,then we cdompare the accuraccy of the model
            self.random_score = accuracy_score(test_y,self.predict_random)
        else:
            self.random_score = roc_auc_score(test_y,self.predict_random)#auc for xgboost
    
        #comparing the two models
        if(self.random_score <  self.xgboost_score):
            return 'XGBoost',self.xgboost
        else:
            return 'RandomForest',self.random_forest
```

```python
import os
class save:
    def __init__(self):
        self.location = "/home/anil/Desktop/waffer_project_weights"
        
    def save_model(self,file_name):
       path = os.path.join(self.location,file_name)
       os.makedirs(path)
        
```

```python
from sklearn.model_selection import train_test_split
for i in list_of_cluster:
    cluster_data = x[x["cluster"]==i]
    cluster_features = cluster_data.drop(["cluster","labels"],axis =1)
    cluster_label = cluster_data["labels"]
    x_train,x_test,y_train,y_test = train_test_split(cluster_features,cluster_label,test_size = 1/3, random_state = 355)
    model_finder = model_finder()#model initilization
    best_model_name,best_model=model_finder.get_best_model(x_train,y_train,x_test,y_test)
    saving = save()
    save_model = saving.save_model(best_model_name+str(i))
```
